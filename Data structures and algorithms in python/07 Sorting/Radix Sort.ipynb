# Radix Sort

**Digit by Digit**

---
###  Least significant digit -> Most significant digit.

---


* Pass1:
> Elements are sorted according to their units(least significant) digits
* Pass2:
> Elements are sorted according to their tens digit.
* Pass3:
> Elements are sorted according to their hundreds digit.\
......................................

* Last Pass:
> Elements are sorted according to their most significant digits.



--- 

* Input data is taken in linked list.
* Queues
* Array Allocation
 > Size of each queue should be n\
 > Space  - n*10\
 > Taking array would __require huge memory__ So linked allocation for queues is needed.

* Linked Allocation (in each pass)
 > Insert nodes into different queues __Rearrange links__\
 > Combine queues back to list __Rearrange links__

```python
class Node:
    def __init__(self,value):
        self.info=value
        self.link=None
        
def radix_sort(start):
    
    rear=[None]*10
    front=[None]*10
    
    leastSigPos=1
    mostSigPos=DigitsInLargest(start)
    
    for k in range(leastSigPos,mostSigPos+1):
        # making all queues empty at beginning of each pass
        for i in range(10):
            rear[i]=None
            front[i]=None
            
        p=start
        while p is not None:
            #find kth digit from right in the number
            dig=Digit(p.info,k)
            
            #Insert the node in Queue(dig)
            if front[dig] is None:
                front[dig]=p
            else:
                rear[dig].link=p
                
            rear[dig]=p
            p=p.link
            
        #join all queues to from new linked list
        
        i=0


def DigitsInLargest(start):
    large=0
    p=start
    while p is not None:
        if p.info>large:
            large=p.info
            
        p=p.link
        
    ndigits=0
    while large!=0:
        ndigits+=1
        large//=10
    return ndigits


start=None
n=int(input("Enter the number of elements:"))

for i in range(n):   #inserting elements in linked list
    data=int(input("Enter element"))
    
    temp=Node(data)
    if start is None:
        start=temp
    else:
        p=start
        while p.link is not None:
            p=p.link
        p.link=temp
        
start=radix_sort(start)
print("Sorted list is:")
p=start
while p is not None:
    print(p.info," ")
    p=p.link
    

print()
```
