````python
def shell_sort(a):
    h=int(input())
    while h>=1:
        for i in range(h,len(a)):
            temp=a[i]
            j=i-h
            while j>=0 and a[j]>temp:
                a[j+h]=a[j]
                j=j-h
                
            a[j+h]=temp
            
        h=h-2
        
a=[1,34,45,23,43,46,4,345,9,66,22,11,55,66,44,34]
shell_sort(a)
print(a)
````

````python
def quick(a):
    sort(a,0,len(a)-1)

    
def sort(a,low,up):
    if low>=up:
        return
    p=partition(a,low,up)
    sort(a,low,p-1)
    sort(a,p+1,up)
    
def partition(a,low,up):
    pivot=a[low]
    i=low+1
    j=up
    
    while i<=j:
        while i<up and a[i]<pivot:
            i+=1
        while a[j]>pivot:
            j=j-1

        if i<j:
            a[i],a[j]=a[j],a[i]
            i+=1
            j-=1
        else:
            break
            
    a[low]=a[j]
    a[j]=pivot
    
    return j
````
