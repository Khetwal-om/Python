* Efficient way is to give list a size

```python
class EmptyStackError(Exception):
    pass
class FullStackError(Exception):
    pass

class Stack:
    def __init__(self,max_length=10):
        self.items=[None]*max_length
        self.count=0
        
    def is_empty(self):
        return self.count==0
    
    def is_full(self):
        return self.count==len(self.items)
    
    def size(self):
        return self.count
    
    def push(self,data):
        if self.is_full():
            raise FullStackError('Stack is full')
        self.items[self.count]=data
        self.count+=1
    
    
    def pop(self):
        if self.is_full():
            raise EmptyStackError('Stack is empty')
        data=self.items[self.count-1]
        self.items[self.count-1]=None
        self.count-=1
        return data
    
    def peek(self):
        if self.is_empty():
            raise EmptyStackError('Stack is empty')
        return self.items[self.count-1]
        
    
    def display(self):
        print(self.items)
        print('-'*25)
        print(self.items[:self.count])
        
        
        
if __name__ == "__main__":
    st=Stack(8)
    
    while True:
        print("1. Push")
        print("2. Pop")
        print("3. Peek")
        print("4. Size")
        print("5. Display")
        print("6. Quit")
        
        
        choice=int(input("Enter your choice"))
        
        
        if choice==1:
            x=int(input("Enter the element to be pushed"))
            st.push(x)
        elif choice==2:
            x=st.pop()
            print("Popped element is",x)
        elif choice==3:
            print("Element at the top is ",st.peek())
            
        elif choice==4:
            print("Size of stack",st.size())
        
        elif choice==5:
            st.display()
        
        elif choice==6:
            break
            
        else:
            print("obselete    Wrong one")
        print()
```

* Linked List implementation of stack

  * Push,pop operation at the end of the list needs traversal of the list. 
     * Push : O(n)
     * Pop  : O(n)
     
  * Push ,pop operation at the beginning of the list needs no traversal of the list.
     * Push : O(1)
     * Pop  : O(1)

```python
class EmptyStackError(Exception):
    pass
class Node:
    def __init__(self,value):
        self.info=value
        self.link=None
        
class Stack:
    
    def __init__(self):
        self.top=None
        
        
    def is_empty(self):
        return self.top==None
    
    def size(self):
        count=0
        if self.is_empty():
            return count
        p=self.top
        while p is not None:
            p=p.link
            count+=1
            
        return count
    
    def peek(self):
        if self.is_empty():
            return
        return self.top.info
    
    
    def push(self,data):
        temp=Node(data)
        temp.link=self.top
        self.top=temp
    
    def pop(self):
        if self.is_empty():
            raise EmptyStackError('no elements to delete')
        popped=self.top.info
        self.top=self.top.link
        return popped
    
    def display(self):
        if self.is_empty():
            return
        p=self.top
        while p is not None:
            print(p.info,'  ',end=' ')
            p=p.link
            
      
        
if __name__ == "__main__":
    st=Stack()
    
    while True:
        print("1. Push")
        print("2. Pop")
        print("3. Peek")
        print("4. Size")
        print("5. Display")
        print("6. Quit")
        
        
        choice=int(input("Enter your choice"))
        
        
        if choice==1:
            x=int(input("Enter the element to be pushed"))
            st.push(x)
        elif choice==2:
            x=st.pop()
            print("Popped element is",x)
        elif choice==3:
            print("Element at the top is ",st.peek())
            
        elif choice==4:
            print("Size of stack",st.size())
        
        elif choice==5:
            st.display()
        
        elif choice==6:
            break
            
        else:
            print("obselete    Wrong one")
        print()
```
