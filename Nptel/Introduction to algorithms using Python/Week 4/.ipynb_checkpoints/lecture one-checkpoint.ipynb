 # O(n**2) sorting algorithms
 
 * Selection sort and insertion sort are both __O(n**2)__ 
 * __O(n**2)__ sorting is infeasible for n over 5000

# Merge Sort

* Sort  A[0:n//2]
* Sort  A[n//2:n]


* Merge sorted halves into B[0:n]

# Divide and Conquer

* Break up problem into disjoint parts.
* Solve each part separately.
* Combine the solutions efficiently.

# Merging sorted lists

* Combine two sorted lists A and B into C

> If A is empty,copy B to C\
> If B is empty ,copy A into c
> Otherwise,compare first element of A and B and move the smaller into C\
> Repeat until all elements in A and B have been moved.

```python
def merge(A,B):
    C=[]
    m=len(A)
    n=len(B)
    i=0
    j=0
    
    while i+j<m+n:
        if i==m:
            C.append(B[j])
            j=j+1
        elif j==n:
            C.append(A[i])
            i=i+1
        elif A[i]<=B[j]:
            C.append(A[i])
            i+=1
        elif A[i]>=B[j]:
            C.append(B[j])
            j+=1
    return C

A=list(range(1,100,2))
B=list(range(2,75,2))
print(merge(A,B))
```
